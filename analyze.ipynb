```python
import pandas as pd
import numpy as np

try:
    dataset = pd.read_csv('new_coursera_reviews.csv', encoding_errors='ignore')
    print(dataset.value_counts)
    print("Number of duplicate reviews:", dataset.duplicated(subset='Review').sum())
    dataset = dataset.drop_duplicates(subset='Review')
except pd.errors.ParserError as e:
    print(f"Error parsing CSV file: {e}")
```

```python
import matplotlib.pyplot as plt

# Assuming 'dataset' is your DataFrame
class_counts = dataset['Rating'].value_counts()

print(dataset.value_counts)

# Plotting the bar graph
class_counts.plot(kind='bar')

# Adding labels and title
plt.xlabel('Class')
plt.ylabel('Count')
plt.title('Composition of Dataset')

# Displaying the graph
plt.show()

import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Load the dataset
# Combine all reviews into a single string
text = ' '.join(review for review in dataset.Review)

# Generate the word cloud
wordcloud = WordCloud(width=800, height=400, background_color='white', max_words=200).generate(text)

# Display the word cloud
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
plt.show()

# Save the word cloud image (optional)
wordcloud.to_file('wordcloud.png')
```
